# Modulation Spectrograms

<b>Overview</b>
<p>  The modulation spectrogram is motivated by the need to represent signals that distinguishes between high frequency carrier waves and the low frequency waves that modulate them. In other words, a given signal might be represented by a general form with some fluctuations along its amplitude. (1)</p>





<b>Spectrogram vs. Modulation Spectrogram  </b>
<p>In many ways, modulation spectrogram is similar to the STFT spectrogram. The spectrogram is a visual 
representation of acoustic frequency vs. time information obtained by taking the square of the magnitude of 
the STFT (1). At a given time, it can tell you what acoustic frequency is present in the signal (with a tradeoff 
between time and frequency resolution, of course). Similarly, given a frequency, it can tell you what time the 
frequency appears. The modulation spectrogram, on the other hand, is a representation of acoustic frequency vs 
modulation frequency. That is, given a modulation frequency, it can tell you what carrier frequencies it modulates 
and given a carrier acoustic frequency, it can tell you which frequencies modulate it.  </p> Note that &quot;acoustic 
frequency&quot; refers to the Fourier decomposition of the signal, and can be seen as the carrier signal. 
The &quot;modulation frequency&quot; is the signal that modulates, or alters the amplitude or frequency of, 
the carrier signal (2). An example of a comparison between the acoustic frequency vs time (STFT spectrogram) 
and the acoustic vs modulation frequency is presented below. It represents two simultaneous speakers, one saying &quot;two&quot; and the other saying &quot;dos.&quot; As you can see, the spectrogram is not able to distinguish between the two sounds. The modulation spectrogram, is able to discriminate between the two speakers and isolate the two fundamental pitches and the harmonics (first and aliased) of the speakers (2). </p>
<img src=shamma_specgram_vs_modspecgram.png>
   

<b>The Model</b>
<p>Presented below is a simplified version of the transformation used to construct the modulation spectrogram 
representation (1).</p>
<img src="shamma_modspecgram_simplified.png">
<p>
The modulation spectrogram is obtained by first selecting the wide frequency bands for the initial filter bank,
usually via a Gabor transform.  This selects the equally spaced channels of desired acoustic frequencies.  
The content in each of these channels is processed via a STFT to get a modulation frequency vs time (frames) 
representation. This is then compressed into a 2D plot of acoustic vs modulation frequency (3).</p>

<b>Sample</b>
<p> We test our sample audio segment using an existing MATLAB toolbox (4) and display the resulting modulation spectrogram. We were not able to find any usable Python implementations for now but are continuing to look into options.</p>

```python
from scipy.io.wavfile import read
import numpy as np
import matplotlib.pyplot as plt

# Params
start_pos = 200000
count_frames = 5000

# Read in the audio
raw_data = read("Cessna 172 - Engine Change Start Up.wav")

data = np.array(raw_data[1][start_pos:start_pos + count_frames, 0], dtype=float) # Only extract one channel

plt.plot(data)
plt.show()
```
