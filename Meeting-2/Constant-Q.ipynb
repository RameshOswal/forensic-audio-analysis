```python
import librosa, librosa.display
import numpy as np
```

```python
import matplotlib.pyplot as plt
X = []
Y = []
for i in range(10):
    filename = "../../heli_and_boat.tar/heli" + str(i) + ".wav"
    y, sr = librosa.load(filename)
    C = librosa.cqt(y, sr=sr)
    X.append(np.ravel(C))
    Y.append(filename.split("/")[-1])
    print("1 read")
    filename = "../../heli_and_boat.tar/boat" + str(i) + ".wav"
    y, sr = librosa.load(filename)
    C = librosa.cqt(y, sr=sr)
    X.append(np.ravel(C))
    Y.append(Y.append(filename.split("/")[-1]))
    print("1 read")
```

```python
X_new = []
for x in X:
    zeros = [0]*(4021668-len(x))
    X_new.append(np.concatenate((x,zeros)))
```

```python
np.savetxt("../../heli_and_boat.tar/X_Cont_Qs.train",X_new,delimiter=",")
```
